```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading ratings file
ratings = pd.read_csv('beer_data.csv', encoding='latin-1')
```

### Data Cleaning

```python
# Getting a summary of missing values.

ratings.isnull().mean()
```

```python
# creating bool series True for NaN values
# since only the review_profilename has NULL values.
bool_series = pd.isnull(ratings["review_profilename"])  
    
# filtering data  
# displaying data only with Review_profilename = NaN  
ratings[bool_series]
```

### Recommendation Models

### Dividing the dataset into train and test

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(filtered_ratings, test_size=0.30, random_state=31)
```

```python
print(train.shape)
print(test.shape)
```

```python
# pivot ratings into movie features
df_beer_features = train.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
).fillna(0)
```

```python
df_beer_features.head()
```

#### Copy train and test dataset
- These dataset will be used for prediction and evaluation.

```python
dummy_train = train.copy()
dummy_test = test.copy()
```

- Dummy train will be used later for prediction of beers which has not been rated by the user. To ignore the beers rated by the user, we will mark it as 0 during prediction. The beers not rated by user is marked as 1 for prediction. 
- Dummy test will be used for evaluation. To evaluate, we will only make prediction on the beers rated by the user. So, this is marked as 1. This is just opposite of dummy_train

```python
dummy_train['review_overall'] = dummy_train['review_overall'].apply(lambda x: 0 if x>=1 else 1)
dummy_test['review_overall'] = dummy_test['review_overall'].apply(lambda x: 1 if x>=1 else 0)
```

```python
# The beers not rated by user is marked as 1 for prediction. 
dummy_train = dummy_train.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
).fillna(1)

# The beers not rated by user is marked as 0 for evaluation. 
dummy_test = dummy_test.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
).fillna(0)
```

```python
dummy_train.head()
```

```python
dummy_test.head()
```

### User Based

#### Using Cosine Similarity

```python
from sklearn.metrics.pairwise import pairwise_distances

# User Similarity Matrix
user_correlation = 1 - pairwise_distances(df_beer_features, metric='cosine')
user_correlation[np.isnan(user_correlation)] = 0
print(user_correlation)
```

```python
user_correlation.shape
```

#### Using adjusted Cosine 

```python
# not removing the NaN values and calculating the mean only for the movies rated by the user
beer_features = train.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
)
```

```python
beer_features.head()
```

##### Normalising the rating of the beerid for each user around 0 mean

```python
mean = np.nanmean(beer_features, axis=1)
df_subtracted = (beer_features.T-mean).T
```

```python
df_subtracted.head()
```

#### Finding cosine similarity

```python
from sklearn.metrics.pairwise import pairwise_distances

# User Similarity Matrix
user_correlation = 1 - pairwise_distances(df_subtracted.fillna(0), metric='cosine')
user_correlation[np.isnan(user_correlation)] = 0
print(user_correlation)
```

#### Prediction

```python
# Ignoring the correlation for values less than 0.
# As we are interested in the users which are more similar to the current users.
user_correlation[user_correlation<0]=0
user_correlation
```

Rating predicted by the user (for beers rated as well as not rated) is the weighted sum of correlation with the beer rating (as present in the filtered_ratings dataset).

```python
user_predicted_ratings = np.dot(user_correlation, beer_features.fillna(0))
user_predicted_ratings
```

```python
user_predicted_ratings.shape
```

Since we are interested only in the beers not rated by the user, we will ignore the beers rated by the user by making it zero.

```python
user_final_rating = np.multiply(user_predicted_ratings,dummy_train)
user_final_rating.head()
```

```python
# Importing matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Determine how similar the first 10 users are to each other and visualise it.
plt.figure(figsize = (30,20))        
sns.heatmap(user_correlation[:10,:10],annot = True)
plt.show()
```

```python
user_correlation[:10,:10]
```

```python
# Finding the top 5 recommendation for the user 05Harley 
user_final_rating.iloc[2].sort_values(ascending=False)[0:5]
```

#### Item Based Similarity

Using Correlation
Taking the transpose of the rating matrix to normalize the rating around the mean for different movie ID. 
In the user based similarity, we had taken mean for each user intead of each beer. 

```python
beer_features = train.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
).T

beer_features.head()
```

Normalising the movie rating for each beer

```python
mean = np.nanmean(beer_features, axis=1)
df_subtracted = (beer_features.T-mean).T
```

```python
df_subtracted.head()
```

Finding the cosine similarity. Since the data is normalised, both the cosine metric and correlation metric will give the same value. 

```python
from sklearn.metrics.pairwise import pairwise_distances

# User Similarity Matrix
item_correlation = 1 - pairwise_distances(df_subtracted.fillna(0), metric='cosine')
item_correlation[np.isnan(item_correlation)] = 0
print(item_correlation)
```

Filtering the correlation only for which the value is greater than 0. (Positively correlated)

```python
item_correlation[item_correlation<0]=0
item_correlation
```

```python
# Determine how similar the first 10 users are to each other and visualise it.
plt.figure(figsize = (30,20))        
sns.heatmap(item_correlation[:10,:10],annot = True)
plt.show()
```

```python
item_correlation[:10,:10]
```

We can convlude the corelation in first 10 beers is higher than that in first 10 users

#### Prediction

```python
item_predicted_ratings = np.dot((beer_features.fillna(0).T),item_correlation)
item_predicted_ratings
```

```python
item_predicted_ratings.shape
```

```python
dummy_train.shape
```

#### Filtering the rating only for the movies not rated by the user for recommendation

```python
item_final_rating = np.multiply(item_predicted_ratings,dummy_train)
item_final_rating.head()
```

```python
# Finding the top 5 recommendation for the user 05Harley
item_final_rating.iloc[2].sort_values(ascending=False)[0:5]
```

#### Evaluation

Will evaluate for the beer already rated by the user insead of predicting it for the beer not rated by the user

##### Using User Similarity

```python
test_beer_features = test.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
)
mean = np.nanmean(test_beer_features, axis=1)
test_df_subtracted = (test_beer_features.T-mean).T

# User Similarity Matrix
test_user_correlation = 1 - pairwise_distances(test_df_subtracted.fillna(0), metric='cosine')
test_user_correlation[np.isnan(test_user_correlation)] = 0
print(test_user_correlation)
```

```python
test_user_correlation[test_user_correlation<0]=0
test_user_predicted_ratings = np.dot(test_user_correlation, test_beer_features.fillna(0))
test_user_predicted_ratings
```

#### Making prediction for the movies rated by the user

```python
test_user_final_rating = np.multiply(test_user_predicted_ratings,dummy_test)
```

```python
test_user_final_rating.head()
```

##### Calculating the RMSE for only the beers rated by user. For RMSE, normalising the rating to (1,5) range. 

```python
from sklearn.preprocessing import MinMaxScaler
from numpy import *

X  = test_user_final_rating.copy() 
X = X[X>0]

scaler = MinMaxScaler(feature_range=(1, 5))
print(scaler.fit(X))
y = (scaler.transform(X))

print(y)
```

```python
test_ = test.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
)
```

```python
# Finding total non-NaN value
total_non_nan = np.count_nonzero(~np.isnan(y))
```

```python
rmse = (sum(sum((test_ - y )**2))/total_non_nan)**0.5
print(rmse)
```

##### Using Item Similarity

```python
test_beer_features = test.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
).T

mean = np.nanmean(test_beer_features, axis=1)
test_df_subtracted = (test_beer_features.T-mean).T

test_item_correlation = 1 - pairwise_distances(test_df_subtracted.fillna(0), metric='cosine')
test_item_correlation[np.isnan(test_item_correlation)] = 0
test_item_correlation[test_item_correlation<0]=0
```

```python
test_item_correlation.shape
```

```python
test_beer_features.shape
```

```python
test_item_predicted_ratings = (np.dot(test_item_correlation, test_beer_features.fillna(0))).T
test_item_final_rating = np.multiply(test_item_predicted_ratings,dummy_test)
test_item_final_rating.head()
```

```python
test_ = test.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
)
```

```python
from sklearn.preprocessing import MinMaxScaler
from numpy import *

X  = test_item_final_rating.copy() 
X = X[X>0]

scaler = MinMaxScaler(feature_range=(1, 5))
print(scaler.fit(X))
y = (scaler.transform(X))


test_ = test.pivot_table(
    index='review_profilename',
    columns='beer_beerid',
    values='review_overall'
)

# Finding total non-NaN value
total_non_nan = np.count_nonzero(~np.isnan(y))
```

#### Finding RMSE

```python
rmse = (sum(sum((test_ - y )**2))/total_non_nan)**0.5
print(rmse)
```

Out of the two models, the performance of `User-based` model is better than the `Item-based` model.
Hence, `User-based` model should be deployed.
